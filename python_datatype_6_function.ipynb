```python
# 1~10까지 합을 리턴하는 함수를 만들고 실행하시오
def su(a):
    result =0
    for i in range(a+1):
        result +=i
    return result
su(10)
```

```python
# 두 개의 입력값 a와 b를 받은 후 a에서 b까지 누적합을 리턴하는 함수
def cumsu(a,b):
    result = 0;
    for i in range(a,b+1):
        result += i
    return result
cumsu(3,5)
```

```python
# 임의의 정수 3개를 매개변수로 입력을 받은 후 ,그 평균을 리턴하는 함수
def avgs(a1,a2,a3):
    result = 0
    sums = a1+a2+a3
    result = sums/3
    return result
avgs(3,5,7)
```

```python
# 키워드 매개변수(**b) : 함수를 호출해서 값을 입력받을 때, 반드시 key=value형태로
#                        값을 입력받는 매개변수, 함수 내에서는 dict형식으로 저장

def test_kwargs(a, **kw):
    print(a)
    print(kw)
    
test_kwargs(1, c = 3, b = 13 )
```

```python
# 임의의 정수 2개 이상을 입력받은 후 짝수의 갯수를 리턴하는 함수
def num_even(*b):
    count = 0
    for i in b:
        if i%2==0:
            count +=1
    return count
num_even(1,3,4,6,78,9,6,4)
```

```python
# 과목과 점수를 매개변수로 입력 받은 후 평균점수를 리턴하는 함수
def sub_avg(**kw):
#     dic={}
#     dic = kw
#     count = 0
#     sums = 0
#     for value in kw.values():
#         count +=1
#         sums +=value
#     return sums/count
    mscore = sum(kw.values())/len(kw)
    return mscore
sub_avg(a=1, b=2, c=7, d=8)
```

```python
# 임의의 정수를 연속 5회 (함수 내에서) 입력받은 후 그 평균을 출력하는 함수..
def con_input():
    sums=0
    for i in range(0,5):
        x=int(input())
        sums+=x
    return sums/5
con_input()
```

```python
# 전역변수와 지역변수..
# 전역변수 - 프로그램 전체에서 호출(사용)가능한 변수
# 지역변수 - 함수 내에서 정의되고 함수 내에서 사용가능한 변수
```

```python
a =100   # 전역변수
def foo():
    global a # 함수 내에서 전역변수를 사용하는 경우 global 키워드 사용
    print(a)
    a=200   # 전역변수 값을 수정
    print(a)
foo()
print(a)    # 변경된 전역변수 출력..
```

```python
# 'stockcode.txt' 파일을 열고 내용을 출력하는 함수..
def showing_file():
    with open('stockcode.txt', 'r') as f:
        for line in f:
            print(line)
showing_file()
```

```python
# 종목명을 입력받은 후 종목코드를 리턴하는 함수..
def code_number():
    with open('stockcode.txt', 'r') as f:
        dic={}
        for line in f:
            val= line.split()[0]
            key=line.split()[1]
            dic[key] = val
        x=str(input())
        for i in dic:
            if i == x:
                return dic.get(x)
        

code_number()
        
```

```python
# 영어문장을 입력받은 후 모음의 갯수를 리턴하는 함수..
def count_vowel(santance : str):
    lists=list(santance)
    check = ['a','i','u','o','e']
    count = 0
    for i in lists:
        if i in check:
            count +=1
    return count
count_vowel('heeeellow wooorldoo')
```

```python
# 내용을 입력하면 메모장 파일에 입력한 내용을 기록하는 함수
def memo(santance : str):
    with open('memo.txt','a') as f:
        f.write(santance)
    with open('memo.txt','r') as f:
        return f.read()
memo('hellow world')
```

```python
# 임의의 정수를 입력받은 후 절대값을 리턴하는 함수. (단, abs()함수는 사용하지 말 것)
def absol(i : int):
    if i>=0:
        return i
    else:
        return i*(-1)
absol(-200)
```

```python
# map
# map(함수, 군집형 데이터) = > 군집형 데이터의 각 요소를 함수에 넣은 결과값을 리턴
# list(map(함수, 군집형데이터)) => 군집데이터 각 요소의 함수 결과값이 리스트에 담김.
# for문을 대체
```

```python
def  tempf(x):
    return x*x
L = [1,2,3,4,5]
```

```python
s = []
for i in L:
    s.append(tempf(i))
print(s)
```

```python
list(map(tempf, L))
```
