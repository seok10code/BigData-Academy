```python
# # 정규표현식

# [] : 대괄화 내의 문자와 일치하는 패턴
# [가나다] : 가 또는 나 또는 다 중에 하나와 일치..
# [가-힣] : 한글
# [0-9] : 숫자
# [a-zA-Z] : 영문자
# [^0-9] : 숫자를 포함하지 않는 패턴
# [^가-힣] : 한글을 포함하지 않는 패턴
    
# \d : 숫자
# \s : 공백문자
# \w : 영문자
    
# ^x : x로 시작하는 패턴
# x$ : x로 끝나는 패턴
# .x : 한 문자와 x로 끝나는 패턴
# x+ : x가 한 번 이상 등장하는 패턴
# x? : x가 0번 또는 한 번
# x* : x가 0번 이상 등장하는 패턴..
# x{n} : x가 n번 등장하는 패턴
# x{n:m} : x가 n번 이상 m번 이하 등장하는 패턴..
```

```python
gym = str(df['공원보유시설(운동시설)'].unique().tolist())
```

```python
gym[:1000]
```

```python
gym = gym.replace('+', ',').replace("'",",").replace(" ",',').replace("/",",")
gym[:1000]
```

```python
re.sub('\d','', gym)
```

```python
gym = re.sub("[\[\d\-nan\.\]]", "", gym)
gym = re.sub(",{2,}", ",", gym)
gym[:1000]
```

```python
gym_count = pd.Series(gym.split(',')).value_counts()
```

```python
plt.figure(figsize= (15,5))
gym_count[:50].plot(kind='bar')
```
