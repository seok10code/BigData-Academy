```python
df_m = df_academy_selected[df_academy_selected['상권업종소분류명'].\
                    isin(['어린이집', '학원-입시'])]
df_m = df_m.sample(1000)   # 임의로 1000개의 데이터(row)
df_m.shape
```

```python
long = df_m.iloc[0].경도
lat = df_m.iloc[0].위도
title = df_m.iloc[0].상호명
address = df_m.iloc[0].도로명주소
```

```python
print(title, lat, long, address)
```

```python
m = folium.Map(location=[lat, long])

folium.Marker(location=[lat, long], tooltip=title, popup='<i>{}</i>'.format(address)).\
add_to(m)

m
```

```python
m.save('index.html')
```

```python
for i in df_m.index[:10]:
    tooltip = df_m.loc[i,'상호명'] + "-" + df_m.loc[i, "도로명주소"]
    lat = df_m.loc[i, '위도']
    long = df_m.loc[i, '경도']
    
    print(tooltip, lat, long)
```

```python
m = folium.Map(location=[lat, long])

for i in df_m.index[:100]:
    tooltip = df_m.loc[i,'상호명'] + "-" + df_m.loc[i, "도로명주소"]
    lat = df_m.loc[i, '위도']
    long = df_m.loc[i, '경도']    
    
    folium.Marker(location=[lat, long], tooltip=tooltip).add_to(m)

m
```
