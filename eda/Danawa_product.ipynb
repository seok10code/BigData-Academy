```python
def get_prod_items(prod_items):  
    
    prod_data = []
    
    for prod_item in prod_items:
        try:
            title = prod_item.select('p.prod_name > a')[0].text.strip()
        except:
            title = ' '
        try:
            spec_list = prod_item.select('div.spec_list')[0].text.strip()
        except:
            spec_list = ' '
        try:
            price = prod_item.select('p.price_sect > a > strong')[0].text.strip()
        except:
            price= ' '

        prod_data.append([title, spec_list, price])
        
    return prod_data
```

```python
# 여러 페이지 크롤링
# => url구조 파악

def get_search_page_url(page):
    return "http://search.danawa.com/dsearch.php?query=%26%2339%3B%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0%20%26%2339%3B&originalQuery=%26%2339%3B%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0%20%26%2339%3B&previousKeyword=%26%2339%3B%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0%20%26%2339%3B&volumeType=allvs&page={}&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&tab=goods".format(page)
```

```python
import tqdm
import time
from tqdm import tqdm_notebook
total_page= 10

for page in tqdm_notebook(range(1, total_page+1)):
    time.sleep(3)
    print(page)
```

```python
driver = webdriver.Chrome('chromedriver.exe')

# page loading 시간을 기다림
time.sleep(3)

total_page = 10    # 크롤링할 총 페이지수
prod_data_total = []   # 각 item정보를 담을 list

for page in tqdm_notebook(range(1, total_page + 1)):
    url = get_search_page_url(page)   # 크롤링할 페이지 url를 return하는 함수
    driver.get(url)  # driver가 url로 페이지 이동
    
    time.sleep(5)  # 페이지 로딩 시간 기다림..
    html = driver.page_source   # 페이지의 html을 다운로드 저장..
    
    # parsing을 위해서 bs가 html읽어서 리턴..
    soup = BeautifulSoup(html, 'html.parser')   
    
    # item  40개를 포함하는 태그범위를 prod_items에 저장
    prod_items = soup.select('div.main_prodlist > ul.product_list > li.prod_item')
    
    # 한 페이지에서 40개의 아이템의 제목, 스펙, 가격을 추출해서 리스에 담는 함수
    prod_item_list = get_prod_items(prod_items)
    
    # 한 페이지의 아이템 정보(제목, 스펙, 가격) 다시 담는 리스트(prod_data_total)
    prod_data_total = prod_data_total + prod_item_list
    
```

```python
use_time_list = []
suction_spec_list = []
category_list = []

for s in df.스펙:
    category = s.split('/')[0]
    for spec in s.split('/')[1:]:
        if '사용시간' in spec:
            use_time = spec.split(':')[1].strip()
        elif '흡입력' in spec:
            suction_spec = spec.split(':')[1].strip()
    category_list.append(category)        
    use_time_list.append(use_time)
    suction_spec_list.append(suction_spec)

len(use_time_list)
df['사용시간'] = use_time_list
df['흡입력'] = suction_spec_list
df['카테고리'] = category
```

```python
df_copy = df[['상품명','가격','사용시간','흡입력']].copy()
df_copy.head()
```

```python
def convert_time_minute(time):
    try:
        if '시간' in time:
            hour = time.split('시간')[0]
            if '분' in time:
                minute = time.split('시간')[-1].split('분')[0]
            else:
                minute = 0
        else:
            hour= 0
            minute = time.split('분')[0]
        return int(hour)*60 +int(minute)
    except:
        None
```

```python
df_copy['사용시간'] = df_copy['사용시간'].apply(convert_time_minute)
```

```python
def get_sucktion(value):
    try:
        value = value.upper()
        if 'AW' in value or 'W' in value:
            result = value.replace('A','').replace('W', '')
            result = int(result.replace(',',''))
        elif 'PA' in value:
            result = value.replace('PA','')
            result = int(result.replace(',',''))/100
        else:
            result = None
        return result
    except:
        return result
    
```

```python
df_copy['흡입력'] = df_copy['흡입력'].apply(get_sucktion)
```
