```python
def kto_df(date):
    df = pd.read_excel('./data/tourist_files/kto_{}.xlsx'.format(date), \
                              header=1,usecols='A:G', skipfooter=4)
    df['기준년월'] = date
    continent_list = ['아시아주','미주','구주','대양주','국적미상','아시아 기타','교포','구주 기타','대양주 기타','미주 기타','아프리카주','기타대륙','교포소계','아프리카 기타']
    country_df = kto_201901[kto_201901['국적'].isin(continent_list)==False]
    country_df.reset_index(drop=True, inplace = True)
    continents = ['아시아']*21 + ['아프리카']*4 + ['유럽']*22 + ['오세아니아']*3 +['아프리카'] + ['기타대륙']*1 + ['교포']
    country_df['대륙'] = continents
    country_df['관광객비율'] = country_df['관광']/country_df['관광'].sum()
    return country_df
```

```python
date='201901'
kto_df(date)
```

```python
kto_dfs = []
for i in range(2010,2020):
    for j in range(1,13):
        try:
            date = str(i)+str(j).zfill(2)
            df = kto_df(date)
            kto_dfs.append(df)
        except:
            pass
        
```
