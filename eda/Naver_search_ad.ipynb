```python
# 중점관리 키워드
# 노출수 상위 95%이면서, 클릭률가 상위 95%인 키워드를 추출..
```

```python
imp = df_index['노출수']
clk = df_index['클릭률']
```

```python
imp[imp > imp.quantile(0.95)]
```

```python
clk[clk > clk.quantile(0.95)]
```

```python
df_index[(imp > imp.quantile(0.80)) & \
         (clk > clk.quantile(0.80))].index
```

```python
# 저효율 키워드 선별..
# 1. 노출수 상위 95% 미만
# 2.클리수 상위 95% 미만
# 3. 초비용 상위 85% 이상..
```

```python
cond1 = df_index['노출수'] < df_index['노출수'].quantile(0.70)
cond2 = df_index['클릭률'] < df_index['클릭률'].quantile(0.7)
cond3 = df_index['총비용(VAT포함,원)'] > df_index['총비용(VAT포함,원)']\
.quantile(0.7)

df_index[cond1 & cond2 & cond3].index
```
