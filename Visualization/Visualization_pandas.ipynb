```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

```python
drinks = pd.read_csv(url)
drinks.head()
```

```python
# drinks.rename(columns = {'col':'col1'})
drinks.columns = ['country', 'beer', 'spirit', 'wine', 'liters', 'continent']
drinks.head()
```

```python
drinks.info()
```

```python
drinks.set_index('country', inplace=True)
```

```python
drinks.head()
```

```python
# 1. Line Chart
```

```python
plt.figure(figsize=(5,3))
drinks['beer'].plot()
```

```python
drinks.plot()
```

```python
plt.figure(figsize=(5,3))
drinks.sort_values('beer')['beer'].plot()
```

```python
# 2. bar chart
```

```python
drinks['beer'].plot(kind='bar')
```

```python
plt.figure(figsize=(11,3))
drinks['beer'].sort_values(ascending = False)[:50].plot(kind='bar')
```

```python
# 대륙별 국가 평균 맥주소비량을 bar char로 표시...
drinks.head(3)
```

```python
plt.figure(figsize=(4,6))
drinks.groupby('continent')['beer'].mean().plot(kind='bar')
```

```python
plt.figure(figsize=(4,6))
drinks.groupby('continent').mean().plot(kind='bar', stacked=True)
```

```python
# histogram - 구간에 포함된 데이터의 수를 bar
```

```python
drinks['beer'].plot(kind='hist', bins = 50)
```

```python
drinks['beer'].plot(kind='hist', bins = 50)
plt.xlabel('Beer Servings')
plt.ylabel('Frequency')
plt.title('Histogram of Beer')
```

```python
# scatter - x좌표, y좌표
```

```python
plt.figure(figsize=(5,3))
drinks.plot(kind='scatter', x='beer', y='wine')
```

```python
pd.plotting.scatter_matrix(drinks)
plt.show()
```

```python
# boxplot
drinks.plot(kind='box')
```
