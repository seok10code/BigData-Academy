```python
plt.figure(figsize=(3,2))

x=[1,2,3,4,5]
y=[i**2 for i in x]

x1=[1,2,3,4,5]
y1=[6,7,8,9,10]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.figure(figsize=(3,2))
x1=range(50)
y1=[np.cos(value) for value in x1]

y2=[np.sin(value) for value in x1]

plt.plot(x1,y1)
plt.plot(x1,y2)
```

```python
plt.figure(figsize=(5,3))

plt.subplot(1,2,1)
plt.plot(x1, y1, color = 'red', linestyle = 'dotted')

plt.subplot(1,2,2)
plt.plot(x1,y2, color = 'green', linestyle = 'dashed')
```

```python
plt.figure(figsize=(6,4))
x1= range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value + 50 for value in x2]

plt.plot(x1, y1, c='#003333', ls='-.', label = 'Line 1')
plt.plot(x2, y2, c='#66FF00', ls=':', label = 'Line 2')

plt.text(50,60, 'Line 1')
plt.annotate('Line 2', xytext=(30, 110), xy = (50, 100),
            arrowprops=dict(facecolor = 'yellow', shrink=0.1))

plt.legend(fancybox=True, shadow=True, loc = 'upper left')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.title('Line Chart')

plt.savefig('lineChart.jpg')
```
