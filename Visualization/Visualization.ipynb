### Line Plot

```python
plt.figure(figsize = (3,2))

x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
```

```python
plt.figure(figsize = (3,2))

x = [1,2,3,4,5]
y = [i**2 for i in x]

x1=[1,2,3,4,5]
y1=[6,7,8,9,10]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.figure(figsize = (3,2))
x1 = range(50)
y1=[np.cos(value) for value in x1]

y2=[np.sin(value) for value in x1]

plt.plot(x1,y1)
plt.plot(x1,y2)
```

```python
plt.figure(figsize=(5,3))

plt.subplot(1,2,1)
plt.plot(x1, y1,color='red', linestyle = 'dotted')

plt.subplot(1,2,2)
plt.plot(x1,y2, color='green', linestyle='dashed')
```

```python
import numpy as np
from scipy.interpolate import splprep, splev

import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.patches import PathPatch

N = 400
t = np.linspace(0, 2 * np.pi, N)
r = 0.5 + np.cos(t)
x, y = r * np.cos(t), r * np.sin(t)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(6,4))
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value + 50 for value in x2]

plt.plot(x1,y1, c='#003333', ls = '-.', label = 'Line 1')      # rgb 색상표
plt.plot(x2,y2, c='#66FF00', ls=':', label = 'Line 2')

plt.text(50,60, 'Line 1')   # 라벨 좌표로 지정해서 만들기
plt.annotate('Line 2', xytext= (30, 110), xy = (50,100), \
             arrowprops=dict(facecolor = 'yellow', shrink=0.1))


plt.legend(fancybox=True, shadow=True, loc = 'upper right')    # 라벨상자 띄우기
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Chart')

plt.savefig('lineChart.jpg')
```

### Bar Plot

```python
y0 = data[0]
y1 = data[1]
y2 = data[2]
```

```python
x = np.arange(0,8,2)
```

```python
x
```

```python
plt.bar(x,y0,color='b',width=0.5)
plt.bar(x+0.5, y1, color='g', width=0.5)
plt.bar(x+1,y2, color='r',width =0.5)
```

```python
plt.bar(x, y0, color='b', width = 0.5)
plt.bar(x, y1, color='g', width=0.5, bottom=np.sum(data[:0], axis=0))
plt.bar(x, y2, color='r', width=0.5, bottom=np.sum(data[:1], axis=0))
```
