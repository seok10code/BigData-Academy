```python
L = [175, 177, 179, 181, 183]
```

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(L)
s
```

```python
# s의 평균, 합, 편차, 분산, 표준편차..
```

```python
d = s.to_frame()
d
```

```python
suma = d.sum()
suma
```

```python
# 평균

mean_val = (suma/len(d)).values
mean_val
```

```python
d['평균'] = mean_val[0]
```

```python
# 편차

d['편차'] = d[0] - mean_val[0]
d
```

```python
# 편차제곱 

d['편차제곱'] = d['편차'].apply(lambda x: x**2)
d
```

```python
# 분산 - 데이터가 중심에서 얼마나 흩어져있는가 (떨어저 있는 정도)

d['분산']= d['편차제곱'].mean()
d
```

```python
# 표준편차 - 데이터가 중심에서 얼마나 흩어져있는가 (떨어저 있는 정도)

d['표준편차'] = d['분산'].apply(lambda x: np.sqrt(x))
d
```

```python
# 상관분석 - 데이터가 어느정도 같은 방향으로 움지이는가.. => "상관계수"
# * 공분산
# * 상관계수
```

```python
# L1과 L2의 상관관계를 분석
L1 = [2,5,4,7,9,5]
L2 = [12,54,86,32,54,72]
```

```python
d = {'L1':[2,5,4,7,9,5],
    'L2':[12,54,86,32,54,72]}
d
```

```python
df = pd.DataFrame(d)
df
```

```python
df['L1'].corr(df['L2'])
```

```python
# 평균계산
df['l1_mean'] = df['L1'].mean()
df['l2_mean'] = df['L2'].mean()
df
```

```python
# 편차계산
df['l1_편차'] = df['L1'] - df['l1_mean']
df['l2_편차'] = df['L2'] - df['l2_mean']
df
```

```python
# 각 편차의 곱
df['ㅣ1_ㅣ2_편차곱'] = df['l1_편차']*df['l2_편차']
df
```

```python
# 편차곱의 합 => 값이 '+' => L1과 L2가 동일한 방향으로 움직인 경향이 더 컸다..
df['ㅣ1_ㅣ2_편차곱'].sum()
```

```python
# 공분산 : 편차곱의 합의 평균 => 함께 움직이는 방향은 알 수 있지만 절대수치값이므로 정도를 파악할 수 없다..
# => 상관계수 => 공분산/(L1표준편차*L2표준편차)
df['공분산'] = df['ㅣ1_ㅣ2_편차곱'].sum()/len(df)
df
```

```python
# 상관계수 => 두 array가 함꼐 움지이는 정도를..
df['L1_std'] = df['L1'].std()
df['L2_std'] = df['L2'].std()
df
```

```python
df['coef'] = df['공분산'] / (df['L1_std']*df['L2_std'])
df
```

```python
# 광고비와 판매액의 각각의 평균, 분산, 표준편차/ 광고비와 판매액의 공분산과 상관계수를 구하시오..
```

```python
import pandas as pd
import numpy as np
```

```python
ad = [13,8,10,15,12,14,15,17]
sales = [94,70,90,100,95,100,85,95]
```

```python
d = {'ad':ad, 'sales':sales}
```

```python
df= pd.DataFrame(d)
df
```

```python
# 광고비와 판매액의 평균

df['ad_mean'] = df['ad'].mean()
df['sales_mean'] = df['sales'].mean()
df
```

```python
# 광고비와 판매액의 편차 
# 편차 : 각 데이터가 평균에서 떨어진 정도
```

```python
df['ad_dev'] = df['ad'] - df['ad_mean']
df['sales_dev'] = df['sales'] - df['sales_mean']
df
```

```python
# ad와 sales의 분산
```

```python
df['ad_var'] = sum(df['ad_dev']**2)/len(df)
df['sales_var'] = sum(df['sales_dev']**2)/len(df)
```

```python
df
```

```python
df['ad'].std(ddof=1)
```

```python
# ad와 sales의 표준편차
```

```python
df['ad_std'] = np.sqrt(df['ad_var'])
df['sales_std'] = np.sqrt(df['sales_var'])
df
```

```python
# ad와 sales의 공분산 => 각각의 편차의 곱의 평균
```

```python
df['ad_sales_cov'] = sum(df['ad_dev']*df['sales_dev'])/len(df)
df
```

```python
# 상관계수..

df['ad_sales_coef'] = df['ad_sales_cov']/(df['ad_std']*df['sales_std'])
df
```

```python
df['ad'].corr(df['sales'])
```

```python
x = np.array([300, 150, 0, -100])
```

```python
Pa = np.array([0.58, 0.87, 0.55, 0.05])
Pb = np.array([0.65, 0.51, 0.45, 0.05])
```

```python
mean_a = sum(x*Pa)
mean_b = sum(x*Pb)
```

```python
print(mean_a, mean_b)  # 선수 a와 b의 상금의 기대값이 높다... => 선수a의 기대값이 더 높다
```

```python
var_a = sum(((x- mean_a)**2)*Pa)
```

```python
var_b = sum(((x- mean_b)**2)*Pb)
```

```python
print(var_a, var_b) # 선수 a, b의 상금의 분산 => 선수a의 분산이 더 크다..
```
