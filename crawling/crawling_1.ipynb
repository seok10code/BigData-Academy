## html
    콘텐츠를 태그로 감싸서 화면을 보여주는 구조 "<", ">"

```python
"""
================================================================
# 하나의 web page

<html>     # 여는 태그 - 가장 상위 태그
    <head>
        web page 기본 정보...
    </head>

    
    
    <body>
        contents..
    </body>

</html>     # 닫는 태그
"""
```

```python
#html test_1.html

<html>     # 여는 태그 - 가장 상위 태그
    <head>
        <title> 파이썬 크롤링 </title>
    </head>

    
    
    <body>
        hello
    </body>

</html>     # 닫는 태그
```

```python
#html test_2.html

<html>     # 여는 태그 - 가장 상위 태그
    <head>
        <title> 파이썬 크롤링 </title>
    </head>

    
    
    <body>
        <table border = 1>
            <tr>
                <td> 항목 </td>
                <td> 2013 </td>
                <td> 2014 </td>
            </tr>
            <tr>
                <td> 매출액 </td>
                <td> 100 </td>
                <td> 200 </td>
            </tr>
            
        </table>
    </body>

</html>     # 닫는 태그
```

```python
# css - html에 효과 부여..
# test_3.html
<html>
    <head>
        <style>
            p {color:blue;}
            div {color: red;}
        </style>
    </head>

    <body>
        <p> paragraph1 </p>   # 문단을 형성해줌 (텍스트처럼)
        <p> paragraph2 </p>
        
        <div> paragraph1</div>   # 내용을 박스안에 담듯 묶음을 만들어줌(블록처럼)
    </body>
</html>
```

```python
# css를 통해서 태그에 효과 입히기..
# test_4.html
<html>
    <head>
        <style>
            #title{color:blue;} # id를 통한 태그에 효과주기 - 전체에 한 곳에만 적용가능
            .par{color: red;}   # class를 통한 태그에 효과주기 - 여러 곳에 적용가능
        </style>
    </head>

    <body>
        <p id = 'title'> paragraph1 </p>     # 'title' id를 적용
        <p class = 'par'> paragraph2 </p>   # 'par' class를 적용
        
        <div class = 'par'> paragraph1</div> # 'par' class를 적용
    </body>
</html>
```
