```python
from bs4 import BeautifulSoup as bs
import requests
url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0'
html = requests.get(url)
soup = bs(html.text, 'html.parser')
path1 = '#clusterResultUL > li > div.wrap_cont > div > div.wrap_tit.mg_tit > a'
res1 = soup.select(path1)
path2 = '#clusterResultUL > li > div.wrap_cont > div > span.f_nb.date'
res2 = soup.select(path2)
path3 = '#clusterResultUL > li > div.wrap_cont > div > p'
res3 = soup.select(path3)
a = len(res1)
for i in range(1,a):
    print('-'*40)
    print('제목 : {}'.format(res1[i].text))
    print('URL : {}'.format(res1[i]['href']))
    print('언론사 : {}'.format(res2[i].text.split('|')[1]))
    print('summary : {}'.format(res3[i].text))
```

```python
from bs4 import BeautifulSoup as bs
import requests
for j in range(1,11):
    path = 'https://movie.naver.com/movie/point/af/list.nhn?&page={}'.format(j)
    url = requests.get(path)
    html = url.text
    soup = bs(html, 'html.parser')
    
    print('==============={} page==============.'.format(j))
    for i in range(10):
        title = soup.select('tr > td > a.movie')[i].text
        star = soup.select('tr > td > div > em')[i].text
        text = soup.select('tr > td > a.report ')[i]['href'].split("'")[5]
        print("{}번 :{}, {}점, 후기 : {} ".format(i+1,title, star, text))
```

```python
def naver_blog_selenium():
    driver = webdriver.Chrome('./chromedriver')
    driver.get('https://www.naver.com')
    elem = driver.find_element_by_id('query')
    elem.send_keys('파이썬')
    elem.send_keys(Keys.RETURN)
    elem = driver.find_element_by_class_name('_blogBase')
    lis = elem.find_elements_by_tag_name('li')
    for li in lis:
        ats = li.find_element_by_class_name('sh_blog_title')
        title = ats.get_attribute()
```
