```python
pop_seoul['외국인비율'] = (pop_seoul['외국인']/pop_seoul['인구수'])*100
pop_seoul.sort_values('외국인비율',ascending=False).head()
```

```python
### CCTV가 가장 많은 구.. 상위 5곳..
```

```python
cctv_seoul.sort_values('소계', ascending=False).head()
```

```python
# 2016년도의 cctv 증가율이 가장 높은 구?
cctv_seoul['2016년 누적'] = cctv_seoul[['2013년도 이전', '2014년', '2015년']].sum(axis=1)
cctv_seoul['2016증가율'] = (cctv_seoul['2016년']/cctv_seoul['2016년 누적'])*100
```

```python
cctv_seoul.sort_values('2016증가율',ascending=False).head(5)
```

```python
# 1인당 cctv수가 (인당 cctv비율)가 가장 높은 상위 5개 구..?
cctv_total= pd.merge(cctv_seoul, pop_seoul, on='구별', how ='inner')
cctv_total.drop(['2013년도 이전','2014년','2015년','2016년','2016년 누적'],axis=1, inplace=True)
cctv_total.head(5)
```

```python
cctv_total['1인당 cctv수'] = (cctv_total['소계']/cctv_total['인구수'])*100
```

```python
# 1인당 cctv 상위 5개 구
cctv_total.sort_values('1인당 cctv수', ascending=False).head(5)
```
